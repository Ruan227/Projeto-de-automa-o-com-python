```python
#leitura da tabela
import pandas as pd
tabela = pd.read_excel("carteira ruan indicacao cartao santander18.01 427 nomes.xlsx")

#cpf variavel   "coluna"
for i, cpf in enumerate(tabela["CPF"]):
    telefone = i, (tabela["TELEFONE 2"])
    #paigina web
    from selenium import webdriver

    navegador = webdriver.Chrome()
    navegador.get("https://www.abrasuacontasnt.com.br/indicacoes")

    #CLICK CARTAO
    navegador.find_element_by_xpath('//*[@id="forme-form"]/div[1]/div[2]/label/span').click()

    #CLICK CORRESPONDENTE 
    navegador.find_element_by_xpath('//*[@id="forme-form"]/div[3]/div[2]/label/span').click()
    #LOJA
    navegador.find_element_by_xpath('//*[@id="codigoCorrespondente"]').send_keys('54254')
    
     #MEU CPF
    navegador.find_element_by_xpath('//*[@id="CPFSocio"]').send_keys('014.794.541-09')
    #CPF DO CLIENTE 
    navegador.find_element_by_xpath('//*[@id="CPF"]').send_keys(cpf)

        #TELEFONE DO CLIENTE 
    navegador.find_element_by_xpath('//*[@id="Telefone"]').send_keys(str(telefone))
       
        
    #TERMOS
    navegador.find_element_by_xpath('//*[@id="AceitaPoliticas"]').click()
```
